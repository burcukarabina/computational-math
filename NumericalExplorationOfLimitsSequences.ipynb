# Numerical Exploration of Limits Through Sequences

In this activity, we will explore what it means for a sequence to **converge**, to **diverge to infinity**, or to **fail to converge**.

The main idea is:

> We do not only care about the first few terms of a sequence.  
> We care about what happens in the **tail** of the sequence.

The **tail** of a sequence means the terms far out in the sequence, after many terms have already passed.

## Part 1: A Sequence That Converges

Consider the sequence

$$
a_n = 1 + \frac{1}{n}.
$$

Let's look at the first 10 terms.

```python
for n in range(1, 11):
    a = 1 + 1/n
    print("a_" + str(n) + " =", a)
```

### Question 1

What value do the terms seem to be getting close to?

Now let's look farther out in the sequence. These are terms from the **tail** of the sequence.

```python
for n in range(50, 61):
    a = 1 + 1/n
    print("a_" + str(n) + " =", a)
```

### Question 2

Do the tail terms support your guess from Question 1?

Explain briefly.

Let's plot the first 100 terms.

```python
import matplotlib.pyplot as plt

n_values = []
a_values = []

for n in range(1, 101):
    a = 1 + 1/n
    n_values.append(n)
    a_values.append(a)

plt.plot(n_values, a_values, marker='o')
plt.axhline(1)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("The sequence a_n = 1 + 1/n")
plt.show()
```

The horizontal line in the graph is at $y=1$.

The terms get closer and closer to this line. We say

$$
\lim_{n\to\infty} \left(1+\frac{1}{n}\right)=1.
$$

## Part 2: A Sequence Can Converge Even If It Oscillates

Now consider

$$
a_n = \frac{(-1)^n}{n}.
$$

This sequence changes sign: positive, negative, positive, negative, and so on.

```python
for n in range(1, 11):
    a = ((-1)**n)/n
    print("a_" + str(n) + " =", a)
```

The sequence oscillates, but the terms are getting closer to 0.

Let's look at the tail.

```python
for n in range(50, 61):
    a = ((-1)**n)/n
    print("a_" + str(n) + " =", a)
```

### Question 3

Even though the signs alternate, what value do the terms seem to approach?

Explain briefly.

```python
n_values = []
a_values = []

for n in range(1, 101):
    a = ((-1)**n)/n
    n_values.append(n)
    a_values.append(a)

plt.plot(n_values, a_values, marker='o')
plt.axhline(0)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("The sequence a_n = (-1)^n / n")
plt.show()
```

This sequence converges to 0:

$$
\lim_{n\to\infty}\frac{(-1)^n}{n}=0.
$$

The signs keep alternating, but the **tail** gets closer and closer to 0.

## Part 3: Divergence to Infinity

Now consider the sequence

$$
a_n = n.
$$

Let's look at some terms.

```python
for n in range(1, 11):
    a = n
    print("a_" + str(n) + " =", a)
```

Now let's look at a later part of the sequence.

```python
for n in range(100, 111):
    a = n
    print("a_" + str(n) + " =", a)
```

This sequence does **not** approach a finite number.

The terms keep getting larger and larger. We say the sequence **diverges to infinity**, and we write

$$
a_n \to \infty.
$$

This does **not** mean the sequence reaches infinity. It means the terms become arbitrarily large.

```python
n_values = []
a_values = []

for n in range(1, 101):
    a = n
    n_values.append(n)
    a_values.append(a)

plt.plot(n_values, a_values, marker='o')
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("The sequence a_n = n")
plt.show()
```

## Part 4: A Sequence That Does Not Converge

Now consider

$$
a_n = (-1)^n.
$$

Let's look at the first several terms.

```python
for n in range(1, 11):
    a = (-1)**n
    print("a_" + str(n) + " =", a)
```

This sequence keeps jumping between $-1$ and $1$.

Let's look at the tail.

```python
for n in range(50, 61):
    a = (-1)**n
    print("a_" + str(n) + " =", a)
```

### Question 4

Does the tail settle down near one number?

Explain why this sequence does not converge.

```python
n_values = []
a_values = []

for n in range(1, 101):
    a = (-1)**n
    n_values.append(n)
    a_values.append(a)

plt.plot(n_values, a_values, marker='o')
plt.axhline(1)
plt.axhline(-1)
plt.xlabel("n")
plt.ylabel("a_n")
plt.title("The sequence a_n = (-1)^n")
plt.show()
```

This sequence does **not** converge.

It also does **not** diverge to infinity.

Instead, it keeps oscillating between two values forever.

## Connection to Function Limits

The same idea will help us understand limits of functions.

For sequences, we ask:

> What happens to $a_n$ when $n$ gets very large?

For functions, we ask:

> What happens to $f(x)$ when $x$ gets close to a certain value?

In both cases, the main idea is that we look at what happens **eventually** or **near the point of interest**, not just at one isolated value.
